```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

def factorial(n):
    return np.sqrt(2*np.pi*n)*((n/math.exp(1))**n)*(1+1.0/(12*n)+1.0/(288*n**2))

def square(alist):
    return [i**2 for i in alist]

l = 1.0           #step length
N = 100           #number of steps taken
n = 5
runs = 10       #number of runs or the number of simulations
#print("PN(n) from theory is " + str((float(math.factorial(N))*(0.5**N))/(math.factorial(n)*math.factorial(N-n))))
finalpos = []     #final position of walker after N steps
bins = 50
histo = np.zeros(bins)
timesteps = np.linspace(0,N,N+1)
Nhits = 0
for j in range(runs):
    xstart = 0.0       #starting point for each run
    x = []             #list of the positions for each run which is rewritten after each run
    x.append(xstart)
    Nleft = 0
    Nright = 0
    for i in range(N):
        a = np.random.uniform(0,1)
        if a < 0.5:
            step = -1.0*l
            Nleft += 1
        else:
            step = 1.0*l
            Nright += 1
        xnew = xstart + step
        x.append(xnew)
        xstart = xnew
    if (x[-1] == n):
        Nhits+=1
    plt.plot(x,timesteps)
    y=np.histogram(x,bins = bins)
    for k in range(bins):
        histo[k] +=y[0][k] 
    finalpos.append(Nright-Nleft)

    #plt.pause(0.05)
    
#print("PN(n) from simulation is "+str(float(Nhits)/N))
#print finalpos
#print 
#print( "<N> = " + str(np.mean(finalpos)), "<d^2> = " + str(np.sqrt(np.mean(square(finalpos))))\
#      , "<N^1/2> = " + str(np.sqrt(N))) 
plt.show()




number = 10
print math.factorial(number)
print factorial(number)
print (math.factorial(number) - factorial(number))/math.factorial(number)


plt.plot(histo/runs)
plt.show()
```
