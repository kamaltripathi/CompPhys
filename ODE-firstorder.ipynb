## This post shows a method of solving a ordinary differential equation. The method used in this tutorial is Euler's forward difference method of discretization.

### equation used is $dx/dt = f(x)$ given some initial condition $x(t = 0)$; the analytical solution to this equation is a decaying exponetial function.


$dx/dt = -x $

or $ dx = dt* (-x)$

$x[i+1] - x[i] = dt* x[i]$


The goal here is to plot the x as a function of t.

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 10
t0, tf = 0,10

###################################################
dt = 0.2
N = int((tf-t0)/dt);
x = np.zeros(N)
t = np.linspace(t0,tf,N)
x[0] = x0

for i in range(N-1):
    x[i+1] = x[i] - dt * x[i]
plt.plot(t,x,'o-', label = f'dt = {dt}')

###################################################

dt = 0.5
N = int((tf-t0)/dt);
x = np.zeros(N)
t = np.linspace(t0,tf,N)
x[0] = x0

for i in range(N-1):
    x[i+1] = x[i] - dt * x[i]
plt.plot(t,x,'o-', label = f'dt = {dt}')

plt.plot(t,x0*np.exp(-t), label = 'Analytical Solution')
plt.xlabel(r'$t$'); plt.ylabel(r'$x$'); plt.legend()
plt.show()
```
